```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
import plotly.io as io
```

```python
#To read data
df = pd.read_csv("Superstore.csv", encoding = "latin-1")
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
#To create additional 3 columns for Month, year and day
df["Order Date"] = pd.to_datetime(df["Order Date"])
df["Ship Date"] = pd.to_datetime(df["Ship Date"])

df["Order Month"] = df["Order Date"].dt.month

df["Order Year"] = df["Order Date"].dt.year

df["Order Day of week"] = df["Order Date"].dt.dayofweek
```

```python
df.head()
```

```python
# To analyse monthly sales
a = df.groupby("Order Month")["Sales"].sum().reset_index()
fig = px.line(a, x = "Order Month", y = "Sales", title = "Monthly Sales Analysis")
fig.show()
```

```python
#To analyse sales as per the category
b = df.groupby("Category")["Sales"].sum().reset_index()
fig = px.pie(b, values = "Sales", names = "Category",hole = 0.5,
             color_discrete_sequence = px.colors.qualitative.Pastel)
fig.update_layout(title_text = "Sales per category")
fig.update_traces(textposition = "inside", textinfo = "percent+label")
fig.show()
```

### **Analysis of Monthly profits**

```python
c = df.groupby("Order Month")["Profit"].sum().reset_index()
fig = px.line(c, x = "Order Month", y = "Profit", title = " Analysis of Monthly profits")
fig.show()
```

**Profit Analysis by Category**

```python
d = df.groupby("Category")["Profit"].sum().reset_index()
fig = px.pie(d, values = "Profit", names = "Category", hole= 0.5 )
fig.update_layout(title_text = "Profit Analysis by category")
fig.update_traces(textposition ="inside", textinfo ="percent+label")
fig.show()
```

**Sales and Profit analysis by Customer segment**

```python
e = df.groupby("Segment").agg({"Sales":"sum", "Profit" : "sum"}).reset_index()
fig = go.Figure()
fig.add_trace(go.Bar(x = e["Segment"], y = e["Sales"], name = "Sales"))
fig.add_trace(go.Bar(x = e["Segment"], y = e["Profit"], name = "Profit"))
fig.update_layout(title_text = "Sales and Profit Analysis by Customer segment")
fig.show()


```
